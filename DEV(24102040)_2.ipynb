1. LINE CHART

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Data
df = pd.read_csv('sales_data.csv')
df['Date'] = pd.to_datetime(df['Date'])

# Line Chart
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Sales'], marker='o', linestyle='-', color='b')
plt.title('Sales Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

* Imports **pandas** to read and handle data from a CSV file.
* Imports **matplotlib** to create the graph.
* Loads the dataset from `sales_data.csv`.
* Converts the Date column into proper date format.
* Creates space for the graph with a specific size.
* Draws a line chart showing Sales over time.
* Adds circle markers and a line to connect the data points.
* Adds a title to the chart.
* Labels the X-axis as Date and Y-axis as Sales.
* Adds grid lines for better readability.
* Rotates the date labels for clear viewing.
* Adjusts layout to avoid overlapping.
* Displays the final chart on the screen.


2. BAR CHART

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Data
df = pd.read_csv('sales_data.csv')

# Bar Chart
category_sales = df.groupby('Category')['Sales'].sum()
plt.figure(figsize=(10, 6))
plt.bar(category_sales.index, category_sales.values, color='skyblue')
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.show()
```

* Groups the data by Category and calculates total sales for each category.

* Draws a bar chart showing total sales for each category.
* Uses sky blue color for the bars.
* Adds a title to the chart.
* Labels the X-axis as Category and Y-axis as Total Sales.
* Displays the bar chart on the screen.

6. AREA PL0T AND STACKED PLOT

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Data
df = pd.read_csv('sales_data.csv')
df['Date'] = pd.to_datetime(df['Date'])
df_pivot = df.pivot_table(values='Sales', index='Date', columns='Category', aggfunc='sum')

# Area Plot
plt.figure(figsize=(10, 6))
df_pivot.plot.area(alpha=0.5)
plt.title('Cumulative Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

* Loads the dataset from `sales_data.csv`.
* Converts the Date column into proper date format.
* Creates a pivot table to organize sales by Date and Category.
* Creates space for the graph with a specific size.
* Draws an area plot to show sales changes over time for each category.
* Uses transparent colors to compare categories easily.
* Adds a title to the chart.
* Labels the X-axis as Date and Y-axis as Sales.
* Displays the area plot on the screen.

9. POLAR CHART

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load Data
df = pd.read_csv('sales_data.csv')
df['Month'] = pd.to_datetime(df['Date']).dt.month
monthly_sales = df.groupby('Month')['Sales'].sum()

# Polar Chart
labels=np.array(monthly_sales.index)
values=np.array(monthly_sales)
angles=np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
values=np.concatenate((values,[values[0]]))
angles+=angles[:1]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, values, color='orange', alpha=0.5)
ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)
plt.title('Monthly Sales Distribution')
plt.show()
```

* Imports **pandas**, **matplotlib**, and **numpy** libraries.
* Loads the dataset from `sales_data.csv`.
* Converts the Date column into month values.
* Groups the data by Month and calculates total sales for each month.
* Stores month labels and sales values in arrays.
* Generates angles to place each month around a circle.
* Repeats the first value and angle to close the circular shape.
* Creates a polar (circular) chart with a specific size.
* Fills the area with orange color and slight transparency.
* Hides radial tick labels for a clean look.
* Sets month numbers around the circle as labels.
* Adds a title to the chart.
* Displays the polar chart on the screen.

11. LOLLIPOP CHART

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Data
df = pd.read_csv('sales_data.csv')
category_sales = df.groupby('Category')['Sales'].sum()

# Lollipop Chart
plt.figure(figsize=(10, 6))
plt.stem(category_sales.index, category_sales.values, basefmt=" ")
plt.title('Sales by Category (Lollipop Chart)')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

* Imports **pandas** to read data from the CSV file.
* Imports **matplotlib** to create the chart.
* Loads the dataset from `sales_data.csv`.
* Groups the data by Category and calculates total sales for each category.
* Creates space for the graph with a specific size.
* Draws a lollipop (stem) chart to show sales for each category.
* Removes the base line for a cleaner look.
* Adds a title to the chart.
* Labels the X-axis as Category and Y-axis as Sales.
* Displays the lollipop chart on the screen.
